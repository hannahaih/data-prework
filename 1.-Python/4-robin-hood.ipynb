## Context
In this challenge, the landing position of arrows shot by archers in the competition will be represented using 2-dimensional coordinates. 

In the 2-dimensional space, a point can be defined by a pair of values that correspond to the horizontal coordinate (x) and the vertical coordinate (y). For example, in our case, an arrow that hits the center of the archery target will land in position (0, 0) on the coordinate axes. 

The space can be divided into 4 zones (quadrants): Q1, Q2, Q3, Q4. If a point is in Q1, both its x coordinate and y coordinate are positive. Any point with a null x or y coordinate is considered to not belong to any quadrant. 

If you want to know more about the cartesian coordinate system, you can check this [link](https://en.wikipedia.org/wiki/Cartesian_coordinate_system).

#### 2. Calculate how many arrows have fallen in each quadrant. 
**Note**: the arrows that fall in the axis (x=0 or y=0) don't belong to any quadrant.

```python
points = [(4, 5), (-0, 2), (4, 7), (1, -3), (3, -2), (4, 5), (3, 2), (5, 7), (-5, 7), (2, 2), (-4, 5), (0, -2),
          (-4, 7), (-1, 3), (-3, 2), (-4, -5), (-3, 2), (5, 7), (5, 7), (2, 2), (9, 9), (-8, -9)]

lst_axis = []
q_one = []
q_two = []
q_three = []
q_four = []

for point in points[0:]:
    if point[0] == 0 or point[1] == 0:
        lst_axis.append(point)
    elif point[0] > 0 and point[1] > 0:
        q_one.append(point)
    elif point[0] < 0 and point[1] > 0:
        q_two.append(point)
    elif point[0] < 0 and point[1] < 0:
        q_three.append(point)
    elif point[0] > 0 and point[1] < 0:
        q_four.append(point)
     

print("\nLISTS OF POINTS")    
print("\n  List of points on axis: ")    
print("   ",lst_axis)

print("\n  List of points in quadrant I:")
print("   ",q_one)

print("\n  List of points in quadrant II:")
print("   ",q_two)

print("\n  List of points in quadrant III:")
print("   ",q_three)

print("\n  List of points in quadrant IV:")
print("   ",q_four)

print("\nCOUNT OF POINTS")

print("\n- Points in quadrant I:", len(q_one))
print("\n- Points in quadrant II:", len(q_two))
print("\n- Points in quadrant III:", len(q_three))
print("\n- Points in quadrant IV:", len(q_four))
```

#### 3. Find the point closest to the center. Calculate its distance to the center. 
Take into account that there might be more than one point at the minimum distance to the center.

**Hint**: Use the Euclidean distance. You can find more information about it [here](https://en.wikipedia.org/wiki/Euclidean_distance).  
**Hint**: Defining a function that calculates the distance to the center can help.

```python
points = [(4, 5), (-0, 2), (4, 7), (1, -3), (3, -2), (4, 5), (3, 2), (5, 7), (-5, 7), (2, 2), (-4, 5), (0, -2),
          (-4, 7), (-1, 3), (-3, 2), (-4, -5), (-3, 2), (5, 7), (5, 7), (2, 2), (9, 9), (-8, -9)]

centerdistance_lst = []

import math

def centerdistance(point):
    a = 0 - float(point[0])
    b = 0 - float(point[1])
    c = math.sqrt((a**2)+(b**2))
    return(c)

print("\nDISTANCE TO CENTER: \n")
    
rdic = {}

for point in points[0:]:
    distance = centerdistance(point)
    centerdistance_lst.append(distance)
    rdic_key = str(point)
    rdic[rdic_key] = distance
    print("   Point",rdic_key,">> Distance:", distance)

print("\n \nLIST OF ALL DISTANCES: \n")
print(centerdistance_lst)



#cd is closest distance
cd = min(rdic.values()) 

def get_point(x):
    for point, distance in rdic.items():
        if distance == x:
            print(point)

print("\n \nCLOSESTS POINTS TO CENTER: \n")
get_point(cd)

```

#### 4. If the archery target has a radius of 9, calculate the number of arrows that won't hit the target. 
**Hint**: Use the function created in step 3. 

```python
points = [(4, 5), (-0, 2), (4, 7), (1, -3), (3, -2), (4, 5), (3, 2), (5, 7), (-5, 7), (2, 2), (-4, 5), (0, -2),
          (-4, 7), (-1, 3), (-3, 2), (-4, -5), (-3, 2), (5, 7), (5, 7), (2, 2), (9, 9), (-8, -9)]

centerdistance_lst = []

import math

def centerdistance(point):
    a = 0 - float(point[0])
    b = 0 - float(point[1])
    c = math.sqrt((a**2)+(b**2))
    return(c)

lst_in = []
lst_out = []
lst_onedge = []

edge = 9

for point in points[0:]:
    if float(centerdistance(point)) == float(edge):
        lst_onedge.append(point)
    elif float(centerdistance(point)) > float(edge):
        lst_out.append(point)
    elif float(centerdistance(point)) < float(edge):
        lst_in.append(point)

print("\nPOINTS INSIDE: \n",len(lst_in))
print("\nPOINTS OUTSIDE: \n",len(lst_out))

if len(lst_onedge) != 0:
    print("\nPOINTS ON EDGE: \n",len(lst_onedge))
```
